# Data Prediction on House Pricing

The goal is to predict the house price based on various features of the house.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('X_train.csv')
print(df.shape)
df.head()
```

```python
y_train = np.log(df.SalePrice) # want log transformation
print(y_train.shape)
y_train.head()
```

```python
X_train = df.drop('SalePrice', axis=1)
print(X_train.shape)
X_train.head()
```

# Model

```python
model = SelectFromModel(Lasso(alpha=0.005, random_state=42))
model.fit(X_train, y_train)

# check for which columns are important
important_features = model.get_support()
print(f'There are {important_features.sum()} important features')
important_features
```

```python
important_features = X_train.columns[important_features]
important_features
```

```python
X_train = X_train[important_features]
X_train.head()
```

Here, we have finalized the extract of the important features, and they are ready to be fed into any machine learning algorithm.
